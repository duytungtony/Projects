# TRAIN SET 

```python
import pandas as pd
import numpy as np
```

```python
# Main dataset
learn_data = pd.read_csv("learn_dataset.csv", index_col=0)

# Job and retirement dataset
data_files = {
    "learn_dataset": "learn_dataset.csv",
    "learn_emp": 'learn_dataset_Emp.csv' ,
    "learn_job": "learn_dataset_job.csv",
    "learn_retired_former": "learn_dataset_retired_former.csv",
    "learn_retired_job": "learn_dataset_retired_jobs.csv",
    "learn_retired_pension": "learn_dataset_retired_pension.csv",
    "learn_sport": "learn_dataset_sport.csv"
}

datasets = {}
for name, path in data_files.items():
    datasets[name] = pd.read_csv(path)

# Geographical and administrative information
city_adm = pd.read_csv("city_adm.csv", encoding="utf-8")
city_loc = pd.read_csv("city_loc.csv", encoding="utf-8")
city_pop = pd.read_csv("city_pop.csv", encoding="utf-8")
departments = pd.read_csv("departments.csv", encoding="utf-8")
regions = pd.read_csv("regions.csv", encoding="utf-8")

na_per_column = learn_data.isna().sum()
print(na_per_column)
```

```python
# Merge job contract data
learn_data = learn_data.merge(datasets["learn_emp"], how="left", on="primary_key")
learn_data
```

```python
na_per_column = learn_data.isna().sum()
print(na_per_column)
```

```python
# Revelant columns from job and retirement to be merged 
learn_job_renamed = datasets["learn_job"].rename(columns={
    'economic_sector': 'job_economic_sector',
    'Job_dep': 'job_Job_dep',
    'job_description': 'job_job_description',
    'type_of_contract': 'job_type_of_contract',
    'JOB_CATEGORY': 'job_JOB_CATEGORY',
    'Work_condition': 'job_Work_condition',
    'COMPANY_CATEGORY': 'job_COMPANY_CATEGORY',
    'Working_hours': 'job_Working_hours',
    'EMPLOYEE_COUNT': 'job_EMPLOYEE_COUNT',
})

learn_retired_job_renamed = datasets["learn_retired_job"].rename(columns={
    'economic_sector': 'retired_economic_sector',
    'Job_dep': 'retired_Job_dep',
    'job_description': 'retired_job_description',
    'type_of_contract': 'retired_type_of_contract',
    'JOB_CATEGORY': 'retired_JOB_CATEGORY',
    'Work_condition': 'retired_Work_condition',
    'COMPANY_CATEGORY': 'retired_COMPANY_CATEGORY',
    'Working_hours': 'retired_Working_hours',
    'EMPLOYEE_COUNT': 'retired_EMPLOYEE_COUNT',  
})

print(learn_job_renamed.columns)
print(learn_retired_job_renamed.columns)
```

```python
learn_data = learn_data.merge(learn_retired_job_renamed, how="left", on="primary_key")
learn_data = learn_data.merge(learn_job_renamed, how="left", on="primary_key")

na_per_column = learn_data.isna().sum()
print(na_per_column)
```

```python
# Fill and replace na of new columns  
learn_data["COMPANY_CATEGORY"] = learn_data["retired_COMPANY_CATEGORY"].fillna(learn_data["job_COMPANY_CATEGORY"])
learn_data["Work_condition"] = learn_data["retired_Work_condition"].fillna(learn_data["job_Work_condition"])
learn_data["economic_sector"] = learn_data["retired_economic_sector"].fillna(learn_data["job_economic_sector"])
learn_data["JOB_CATEGORY"] = learn_data["retired_JOB_CATEGORY"].fillna(learn_data["job_JOB_CATEGORY"])
learn_data["Working_hours"] = learn_data["retired_Working_hours"].fillna(learn_data["job_Working_hours"])
learn_data["EMPLOYEE_COUNT"] = learn_data["retired_EMPLOYEE_COUNT"].fillna(learn_data["job_EMPLOYEE_COUNT"])
learn_data["job_desc"] = learn_data["retired_job_description"].fillna(learn_data["job_job_description"])
learn_data["Job_dep"] = learn_data["retired_Job_dep"].fillna(learn_data["job_Job_dep"])
learn_data["type_of_contract"] = learn_data["retired_type_of_contract"].fillna(learn_data["job_type_of_contract"])

# Drop the redundant columns after merging
learn_data.drop(columns=[
    'retired_COMPANY_CATEGORY', 'job_COMPANY_CATEGORY', 
    'retired_Work_condition', 'job_Work_condition', 
    'retired_economic_sector', 'job_economic_sector', 
    'retired_JOB_CATEGORY', 'job_JOB_CATEGORY',
    'retired_Working_hours', 'job_Working_hours', 
    'retired_EMPLOYEE_COUNT', 'job_EMPLOYEE_COUNT', 
    'retired_job_description', 'job_job_description', 
    'retired_Job_dep', 'job_Job_dep', 
    'retired_type_of_contract', 'job_type_of_contract'
], inplace=True)

learn_data
```

```python
# Same total number of missings as when merging separately (job and retirement)
na_per_column = learn_data.isna().sum()
print(na_per_column)
```

```python
learn_data.columns
```

```python
learn_data = learn_data.merge(datasets["learn_retired_former"], how="left", on="primary_key")
learn_data = learn_data.merge(datasets["learn_retired_pension"], how="left", on="primary_key")
learn_data = learn_data.merge(datasets["learn_sport"], how="left", on="primary_key")

learn_data 
```

```python
na_per_column = learn_data.isna().sum()
print(na_per_column)
```

```python
learn_data["Insee_code"] = learn_data["Insee_code"].astype(str)

# Merge learn_data with complementary file on dep (2 first characters of Insee_code)
learn_data["dep"] = learn_data["Insee_code"].str[:2]

learn_data = learn_data.merge(departments, how="left", left_on='dep', right_on='DEP').drop(columns=['dep'])
learn_data = learn_data.merge(regions, how="left", on="REG")
learn_data = learn_data.merge(city_loc, how="left", on="Insee_code")
learn_data = learn_data.merge(city_adm, how="left", on="Insee_code")
learn_data = learn_data.merge(city_pop, how="left", on="Insee_code")

learn_data
```

```python
learn_data.columns
```

```python
na_per_column = learn_data.isna().sum()
print(na_per_column)
```

```python
# Fill missing numerical values with 0 (e.g., no pension)
numerical_cols = learn_data.select_dtypes(include=["float", "int"]).columns
learn_data[numerical_cols] = learn_data[numerical_cols].fillna(0)

# Fill missing categorical values with "Unknown"
categorical_cols = learn_data.select_dtypes(include=["object"]).columns
learn_data[categorical_cols] = learn_data[categorical_cols].fillna("Unknown")
```

# MODEL AND EVALUATION

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

X = learn_data.drop(columns=['target'])
y = learn_data['target']

# Lable Encoding to categorical features - after splitting to avoid data leakage 
label_encoder = LabelEncoder()
categorical_columns = X.select_dtypes(include=['object', 'category']).columns
            
for col in categorical_columns:
    X[col] = label_encoder.fit_transform(X[col])

# Split dataset 
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)
```

## LINEAR

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

y_val_pred_l = linear_model.predict(X_val)

# Evaluate 
mse_linear = mean_squared_error(y_val, y_val_pred_l)
r2_linear = r2_score(y_val, y_val_pred_l)

print(f"Linear Regression MSE: {mse_linear}")
print(f"Linear Regression R²: {r2_linear}")
```

## RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Train the Random Forest Regressor
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)

# Evaluate the model using MSE and R²
y_val_pred_rf = rf.predict(X_val)
mse = mean_squared_error(y_val, y_val_pred_rf)
r2 = r2_score(y_val, y_val_pred_rf)

print("Random Forest MSE:", mse)
print("Random Forest R²:", r2)
```

### Grid Search for Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Optimize meta-params with Cross-Validation
rf_grid = {
    'min_samples_split': [2, 5, 10],
    'max_depth': [10, 20]
}

rf_grid_search = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid=rf_grid,
    scoring='neg_mean_squared_error', 
    cv=5,  
    n_jobs=-1
)
rf_grid_search.fit(X_train, y_train)

# Future Performance
best_params = rf_grid_search.best_params_
best_neg_mse = rf_grid_search.best_score_ 
future_performance_mse = -best_neg_mse  

print("Best Parameters:", best_params)
print("Future Performance (Mean CV MSE):", future_performance_mse)

# Performance on unseen data (validation dataset X_val, y_val)
best_model = rf_grid_search.best_estimator_
y_val_pred = best_model.predict(X_val)

mse_val = mean_squared_error(y_val, y_val_pred)
r2_val = r2_score(y_val, y_val_pred)

print("Validation MSE (Supplementary):", mse_val)
print("Validation R² (Supplementary):", r2_val)
```

## XG BOOST

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# GridSearch
xgb_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 200],
    'subsample': [0.8, 1.0]
}

xgb_grid_search = GridSearchCV(
    XGBRegressor(objective='reg:squarederror', random_state=42),
    param_grid=xgb_grid,
    scoring='neg_mean_squared_error',  
    cv=5,  
    n_jobs=-1,
    verbose=1
)

xgb_grid_search.fit(X_train, y_train)

# Future Performance: Mean MSE từ Cross-Validation
best_params = xgb_grid_search.best_params_
best_neg_mse = xgb_grid_search.best_score_  
future_performance_mse = -best_neg_mse  

print("Best Parameters:", best_params)
print("Future Performance (Mean CV MSE):", future_performance_mse)

# Performance on unseen data (validation dataset X_val, y_val)
best_model = xgb_grid_search.best_estimator_
y_val_pred = best_model.predict(X_val)

val_mse = mean_squared_error(y_val, y_val_pred)
val_r2 = r2_score(y_val, y_val_pred)

print("Validation MSE (Supplementary):", val_mse)
print("Validation R² (Supplementary):", val_r2)
```
